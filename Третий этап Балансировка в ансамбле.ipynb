```python
# Загрузка библиотек
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# загрузка сабмиссий для ансамблевой балансировки
# ноутбук, получающий сабмиссию 13 с результатом 0.788480 на лидерборде
# приведён в образцовом ноутбуке этапа 3 и полностью воспроизводим
sub1 = pd.read_csv('submission_1.csv', sep=',')
sub2 = pd.read_csv('submission_2.csv', sep=',')
sub3 = pd.read_csv('submission_3.csv', sep=',')

sub4 = pd.read_csv('submission_4.csv', sep=',')
sub5 = pd.read_csv('submission_5.csv', sep=',')
sub6 = pd.read_csv('submission_6.csv', sep=',')

sub7 = pd.read_csv('submission_7.csv', sep=',')
sub8 = pd.read_csv('submission_8.csv', sep=',')
sub9 = pd.read_csv('submission_9.csv', sep=',')

sub10 = pd.read_csv('submission_10.csv', sep=',')
sub10.columns=['ID','Статус_1']
sub11 = pd.read_csv('submission_11.csv', sep=',')
sub11.columns=['ID','Статус_2']
sub12 = pd.read_csv('submission_12.csv', sep=',')
sub12.columns=['ID','Статус_3']
# сабмиссия, приведённая и воспроизводимая в образцовом ноутбуке
sub13 = pd.read_csv('submission_13.csv', sep=',')
sub13.columns=['ID','Статус_4']

```

```python
# объединение во фреймы по группам
data_frames_1 = [sub1, sub2, sub3]
data_frames_2 = [sub4, sub5, sub6]
data_frames_3 = [sub7, sub8, sub9]
data_frames_4 = [sub10, sub11, sub12, sub13]
```

```python
# создание сводной таблицы прогнозов по каждой группе
df_merged_1 = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],how='outer'), data_frames_1)
df_merged_2 = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],how='outer'), data_frames_2)
df_merged_3 = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],how='outer'), data_frames_3)
df_merged_4 = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],how='outer'), data_frames_4)
```

```python
# модирование
df_mode_1=df_merged_1.mode(axis=1) 
df_mode_2=df_merged_2.mode(axis=1) 
df_mode_3=df_merged_3.mode(axis=1) 
df_mode_4=df_merged_4.mode(axis=1) 
```

```python
# показатели по группам по результатам модирования
df_mode_1[0].value_counts()
```

```python
df_result_1=df_merged_1.join(df_mode_1)
df_result_1['Сумма']=df_result_1['Статус_x']+df_result_1['Статус_y']+df_result_1['Статус']

df_result_2=df_merged_2.join(df_mode_2)
df_result_2['Сумма']=df_result_2['Статус_x']+df_result_2['Статус_y']+df_result_2['Статус']

df_result_3=df_merged_3.join(df_mode_3)
df_result_3['Сумма']=df_result_3['Статус_x']+df_result_3['Статус_y']+df_result_3['Статус']

df_result_4=df_merged_4.join(df_mode_4)
df_result_4['Сумма']=df_result_4['Статус_1']+df_result_4['Статус_2']+df_result_4['Статус_3']+df_result_4['Статус_4']


```

```python
df_result_1['Статус']=df_result_1['Статус'].astype(int)
df_result_2['Статус']=df_result_2['Статус'].astype(int)
df_result_3['Статус']=df_result_3['Статус'].astype(int)
df_result_4['Статус']=df_result_4['Статус'].astype(int)
```
